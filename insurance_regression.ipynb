```python
# import the relevant packages

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns

from sklearn.linear_model import LinearRegression 


```

```python
# make a scatter plot with the insurance.csv data

ins_data = pd.read_csv('insurance.csv')
x = ins_data['age'].values.reshape(-1, 1)
y = ins_data['charges'].values.reshape(-1, 1)
plt.scatter(x, y, s=5)

# make a prediction function using a linear regression model
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)


```

```python
# plot the scatter plot and prediction function 

plt.scatter(x, y, s=5)
plt.plot(x, y_pred) 
plt.title('Impact of age on Charges - Regression Line')
plt.xlabel('Age')
plt.ylabel('Charge')


```
